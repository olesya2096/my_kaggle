```python
data = data.drop(['store_sales(in millions)', 'unit_sales(in millions)', 'id', 'gross_weight', 'units_per_case', 'recyclable_package', 'low_fat'], axis=1)
```

```python
!pip install catboost
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
```

```python
X = data.drop(['cost'], axis=1)
y = data.cost
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
cat = CatBoostRegressor(iterations = 500,
    loss_function = 'RMSE',
    random_state = 42,
    bootstrap_type = 'Bayesian', 
    depth = 12,
    l2_leaf_reg = 3,
    learning_rate = 0.08)

cat.fit(X_train, y_train)
```

```python
y_pred_train = cat.predict(X_train)
y_pred_test = cat.predict(X_test)

score_train = mean_squared_error(y_train, y_pred_train)
score_test = mean_squared_error(y_test, y_pred_test)
print(f'train error: {score_train}')
print(f'test error: {score_test}')
```

```python
feature_importance = cat.feature_importances_
sorted_f = np.argsort(feature_importance)
fig = plt.figure(figsize=(10, 5))
plt.barh(range(len(sorted_f)), feature_importance[sorted_f], align='center')
plt.yticks(range(len(sorted_f)), np.array(X_test.columns)[sorted_f])
plt.title('Feature Importance')
```
